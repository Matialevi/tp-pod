# Yahoo! Finance

Vamos a analizar acciones. La siguiente linea accede a Yahoo Finance y devuelve un DataFrame con los valores de la acción cada dia desde el 1980.

```python
import yfinance as yf
dataframe = yf.download('AAPL', start="1980-01-01", end="2030-01-01")
```

¿Cual es el registro mas viejo? Imprimirlo.

```python
df= dataframe
df.head(1)
```

Cree la columna `Average` tal que

$$Average =  \frac{High-Low}{2}$$

y muestre con un histograma dicha columna.

```python
df['Average']= df['High']-df['Low']/2
df['Average'].hist()
```

Con `matplotlib`, muestre como `Average` fue evolucionando *al final de cada año*.

```python
import matplotlib.pyplot as plt
df.resample('Y').last()['Average'].plot()
```

Muestre con un gráfico de barras, como el volumen fue cambiando *año a año*.

```python
# Asegurarse de que los datos de la fecha estén en formato de datetime
dataframe.index = pd.to_datetime(dataframe.index)

# Agregar una columna para el año
dataframe['Year'] = dataframe.index.year

# Agrupar los datos por año y sumar el volumen para cada año
volumen_por_year = dataframe.groupby('Year')['Volume'].sum()

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
volumen_por_year.plot(kind='bar', color='skyblue')
plt.title('Volumen de AAPL año a año')
plt.xlabel('Año')
plt.ylabel('Volumen')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
# Cargar los archivos CSV en DataFrames
airlines = pd.read_csv('airlines.csv')
airports = pd.read_csv('airports.csv')
flights = pd.read_csv('flights.csv', low_memory=False)

# Verificar los datos
airlines.head(),airports.head(),flights.head()
```

Combine (*join*) las tablas `airlines`, `airports`, y `flights` en una sola tabla.

```python
# Combinar flights con airports utilizando 'ORIGIN_AIRPORT' y 'IATA_CODE' para el aeropuerto de origen
vuelos_combinados = pd.merge(flights, airports, how='left', left_on='ORIGIN_AIRPORT', right_on='IATA_CODE')

# Combinar flights con airlines utilizando 'AIRLINE' y 'IATA_CODE'
vuelos_combinados = pd.merge(vuelos_combinados, airlines, how='left', left_on='AIRLINE', right_on='IATA_CODE')

# Verificar las primeras filas de la tabla combinada
print(vuelos_combinados.head())
```

¿Cuantos vuelos fueron al aeropuerto JFK?

```python
destino_jfk = vuelos_combinados[vuelos_combinados['AIRPORT'] == 'John F. Kennedy International Airport']
print(len(destino_jfk))
```

¿Cuantos vuelos hizo la aerolinea AA?

```python
vuelos_aerolinea_AA= vuelos_combinados[vuelos_combinados['AIRLINE_x'] == 'AA']
print(len(vuelos_aerolinea_AA))
```

¿Que aerolineas (las primeras 5) tuvo la menor cantidad de vuelos con atrasos? Imprimirlas.

```python
aerolineas_con_retrasos_menor = vuelos_combinados[vuelos_combinados['ARRIVAL_DELAY'] < 0]['AIRLINE_x'].value_counts().head(5)
print(aerolineas_con_retrasos_menor)
```

¿Que aerolineas (las primeras 5) tuvo la mayor cantidad de vuelos con atrasos? Imprimirlas.

```python
aerolineas_con_mayor_retrasos = vuelos_combinados[vuelos_combinados['ARRIVAL_DELAY'] > 0]['AIRLINE_x'].value_counts().head(5)
print(aerolineas_con_mayor_retrasos)
```
